#Listas
Es una colección ordenada y mutable de elementos. Son de una dimensión y el orden es importante.

```python
#Listas
numeros = [50, 40, 50, 60, 56]
cursos = ["Historia", "Economía", "Derecho", "TEIT"]
mi_lista = [10, "PUCP", True]
print(numeros)
print(cursos)
print(mi_lista)
print(type(cursos))

#Lista vacía
vacia = []
print(len(vacia))
```

```python
#len: nos da la longitud de un objeto
len(cursos)
```

```python
#Operaciones con listas: Concatenar
nueva_lista = cursos + mi_lista
print(nueva_lista)

#El orden es importante y empieza en 0:
print(nueva_lista[2])

#Multiplicación: se usa para repetir la lista el número de veces que se desee.
lista_repetida = cursos * 2
print(lista_repetida)

#Al imprimir un rango de la lista, se toma el objeto anterior al rango.
print(lista_repetida[2:5])

#Cuando se usan indices negativos se empieza a contar desde atrás
print(lista_repetida[-1])

# La lista 'lista_repetida' tiene 8 elementos (índices 0 a 7).
print(lista_repetida[5:11])
print(lista_repetida[5:7])
print(lista_repetida[len(lista_repetida)-2:len(lista_repetida)])
```

```python
number = [1, 2, 3, 4, 5, 6, 7]
print(number[0:4:3])
```

```python
#Las listas se pueden modificar
print(lista_repetida)
lista_repetida[2] = "Matemáticas"
print(lista_repetida)
```

```python
#Pequeño experimento
lista_1 = [1, 2, 3, 4, 5]
lista_2 = lista_1 #Lista 2 se le está asignando el OBJETO, no el valor de la lista 1
print(lista_1)
print(lista_2)

lista_1[3] = 20
print(lista_1)
print(lista_2)

#Qué puedo hacer para que el cambio en la lista 1 no afecte a la lista 2?
lista_2 = lista_1.copy() #Con copy() copiamos en VALOR de la lista, mas no el OBJETO
lista_1[3] = 30
print(lista_1)
print(lista_2)

```

##Métodos de las listas

```python
print(dir(list))
```

### `append()`
Agrega un elemento al final de la lista.

```python
mi_lista = [1, 2, 3]
mi_lista.append(4)
print(f"Después de append(4): {mi_lista}")
```

### `clear()`
Elimina todos los elementos de la lista.

```python
mi_lista = [1, 2, 3]
mi_lista.clear()
print(f"Después de clear(): {mi_lista}")
```

### `copy()`
Devuelve una copia superficial de la lista.

```python
mi_lista_original = [1, 2, 3]
mi_lista_copia = mi_lista_original.copy()
mi_lista_original.append(4)
print(f"Lista original: {mi_lista_original}")
print(f"Lista copia (no afectada): {mi_lista_copia}")
```

### `count()`
Devuelve el número de veces que aparece un valor especificado en la lista.

```python
mi_lista = [1, 2, 2, 3, 2]
conteo_dos = mi_lista.count(2)
print(f"El número 2 aparece {conteo_dos} veces.")
```

### `extend()`
Agrega los elementos de un iterable (como otra lista) al final de la lista actual.

```python
mi_lista = [1, 2, 3]
otra_lista = [4, 5]
mi_lista.extend(otra_lista)
print(f"Después de extend([4, 5]): {mi_lista}")
```

### `index()`
Devuelve el índice de la primera aparición del valor especificado.

```python
mi_lista = ['manzana', 'banana', 'cereza']
indice_banana = mi_lista.index('banana')
print(f"'banana' está en el índice: {indice_banana}")
```

### `insert()`
Agrega un elemento en la posición especificada.

```python
mi_lista_insert = ['a', 'b', 'd']
mi_lista_insert.insert(2, 'c')
print(f"Después de insert(2, 'c'): {mi_lista_insert}")
```

### `pop()`
Elimina y devuelve el elemento en el índice especificado (o el último elemento si no se especifica el índice).

```python
mi_lista = [10, 20, 30, 40]
elemento_eliminado = mi_lista.pop(1)
print(f"Elemento eliminado con pop(1): {elemento_eliminado}")
print(f"Lista después de pop(1): {mi_lista}") #El pop funciona como una pila en el sentido de apilar
```

### `remove()`
Elimina la primera aparición del valor especificado.

```python
mi_lista = ['rojo', 'verde', 'azul', 'verde']
mi_lista.remove('verde')
print(f"Después de remove('verde'): {mi_lista}")
```

### `reverse()`
Invierte el orden de los elementos de la lista en su lugar.

```python
mi_lista = [5, 1, 4, 2, 8]
mi_lista.reverse()
print(f"Después de reverse(): {mi_lista}")
```

### `sort()`
Ordena los elementos de la lista en su lugar (por defecto, en orden ascendente).

```python
mi_lista_sort = [5, 1, 4, 2, 8]
mi_lista_sort.sort()
print(f"Después de sort(): {mi_lista_sort}")

mi_lista_sort_desc = [5, 1, 4, 2, 8]
mi_lista_sort_desc.sort(reverse=True)
print(f"Después de sort(reverse=True): {mi_lista_sort_desc}")
```
